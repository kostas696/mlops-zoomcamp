To execute the entire training pipeline with Prefect and MLflow using the March 2023 NYC Yellow Taxi dataset, I ran the following command in the terminal:

python train_flow.py --year 2023 --month 3

This triggered the following steps:

- Automatically downloaded the dataset if not found locally

- Loaded and preprocessed 3,403,766 records

- Filtered valid trips to 3,316,216 entries

- Trained a linear regression model

- Logged the model and metadata to MLflow

- Computed and printed the intercept and model size
